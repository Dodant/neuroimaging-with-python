```python
def display_images(fixed_image_z, moving_image_z, fixed_npa, moving_npa):
    plt.subplots(1,2,figsize=(10,8))
    plt.subplot(1,2,1)
    plt.imshow(fixed_npa[:,:,fixed_image_z],cmap=plt.cm.Greys_r);
    plt.title('fixed image')
    plt.axis('off')
    plt.subplot(1,2,2)
    plt.imshow(moving_npa[:,:,moving_image_z],cmap=plt.cm.Greys_r);
    plt.title('moving image')
    plt.axis('off')
    
    plt.show()
```

```python
nb_features = [
    [32, 32, 32, 32],         # encoder features
    [32, 32, 32, 32, 32, 16]  # decoder features
]

# Preprocessing
input_img = 'PKR_T2.nii'
temp_img = 'voxelmorph\\templates\\mni_icbm152_t2_sym_09c_rss.nii'

prk_t2 = nib.load(input_img).get_fdata()
prk_t2 = (prk_t2 - prk_t2.min()) / (prk_t2.max() - prk_t2.min())
prk_t2 = np.rot90(prk_t2)

temp_t2 = nib.load(temp_img).get_fdata()
temp_t2 = (temp_t2 - temp_t2.min()) / (temp_t2.max() - temp_t2.min())
temp_t2 = np.rot90(temp_t2)

# shape = 192 / 160 / 224
```

```python
nb_vis = 5

idx = np.random.choice(prk_t2.shape[2], nb_vis, replace=False)
example_prk = [prk_t2[:,:,f] for f in idx]
example_tem = [temp_t2[:,:,f] for f in idx]

ne.plot.slices(example_prk, cmaps=['gray'], do_colorbars=True, titles=idx)
ne.plot.slices(example_tem, cmaps=['gray'], do_colorbars=True, titles=idx)
```

```python
interact(display_images, 
         fixed_image_z=(0, temp_t2.shape[2]-1), 
         moving_image_z=(0, prk_t2.shape[2]-1), 
         fixed_npa=fixed(temp_t2),
         moving_npa=fixed(prk_t2))
```

```python
vol_shape = prk_t2.shape[1:]
vxm_model = vxm.networks.VxmDense(vol_shape, nb_features, int_steps=0)
vxm_model.load_weights('brain_2d_smooth.h5')
example_prk_1 = example_prk[0]
example_prd_1 = vxm_model.predict(example_prk_1)
```
